```python
class Car:
    def __init__(self, make, model, price):
        self.make = make
        self.model = model
        # Task 2: Private attribute
        self.__price = float(price)

    # Task 2: Getter method
    def get_price(self):
        return self.__price

    # Task 2: Setter method
    def set_price(self, new_price):
        if new_price >= 0:
            self.__price = float(new_price)
        else:
            print("Error: Price cannot be negative.")

    def apply_discount(self, discount_percentage):
        discount_amount = self.__price * (discount_percentage / 100)
        self.__price -= discount_amount
        print(f"Applied {discount_percentage}% discount. New price: ${self.__price:,.2f}")

    def get_info(self):
        return f"Make: {self.make}, Model: {self.model}, Price: ${self.__price:,.2f}"

# --- Demonstration of Tasks 1 & 2 ---
my_car = Car("Toyota", "Corolla", 25000)

# Demonstrating Encapsulation
print(my_car.get_info())
my_car.set_price(23000)  # Using setter
print(f"Updated Price via getter: ${my_car.get_price()}")
```

```python
class ElectricCar(Car):
    def __init__(self, make, model, price, battery_range):
        # Call the parent constructor
        super().__init__(make, model, price)
        self.battery_range = float(battery_range)

    # Overriding get_info
    def get_info(self):
        base_info = super().get_info()
        return f"{base_info}, Battery Range: {self.battery_range} miles"

class SportsCar(Car):
    def __init__(self, make, model, price, top_speed):
        super().__init__(make, model, price)
        self.top_speed = float(top_speed)

    # Overriding get_info
    def get_info(self):
        base_info = super().get_info()
        return f"{base_info}, Top Speed: {self.top_speed} mph"
```

```python
# Create a list of different Car-type objects
fleet = [
    Car("Ford", "Mustang", 35000),
    ElectricCar("Tesla", "Model 3", 45000, 350),
    SportsCar("Ferrari", "488 GTB", 250000, 205)
]

print("\n--- Displaying Fleet Information (Polymorphism) ---")
for car in fleet:
    # Even though objects are different classes, they all share the get_info() method
    print(car.get_info())
```
